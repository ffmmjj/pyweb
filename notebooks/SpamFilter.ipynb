```python
def parse_message(html):
    dom = web.Element(html)
    divs = dom('div')
    result = defaultdict(str)
    
    email = [{div('label')[0].content: div('span')[0].content} for div in divs if len(div('label')) > 0 and len(div('span')) > 0]
    subject = [{div('label')[0].content: div('div')[0].content} for div in divs if len(div('div')) > 0]

    [email.append(s) for s in subject]

    for field in email:      
        for k,v in field.iteritems():
            result[k] = v
    
    return result

parse_message(html)
```

```python
def parse_content(message):
    result = ""
    try:
        result = parse_message(base64.b64decode(message))[u'Mensagem:'].replace(',','').lower().strip().encode('ascii','ignore')
    except:
        pass
    return result
```

```python
def create_words_dict(messages):
    
    bag = defaultdict(int)
    for m in messages:
        try:
            content = parse_content(m)
            for w in content.split():
                bag[w] += 1
        except:
            pass

    sorted_bag = sorted(bag.items(), key=operator.itemgetter(1), reverse=True)

    return sorted_bag
```

```python
word_dict = create_words_dict(messages)
```

```python
word_dict = [w for w in word_dict if w[1] >= 100 and len(w[0]) > 3]
print 'word count: %d' % len(word_dict)

for item in word_dict:
    word = item[0]
    count = item[1]
    print word + ' -> ' + str(count)
```

```python
def create_features(word_dict, messages):
    words = [item[0] for item in word_dict]
    features = list()
    for m in messages:
        feature = list()
        words_in_message = parse_content(m).split()
        for w in words:                    
            if w in words_in_message:
                feature.append(1)
            else:
                feature.append(0)
        features.append(feature)
    return features
    
features = create_features(word_dict, messages)
print features[-1]
```

```python
training_samples = 1000
X = features[:training_samples]
y = [random.choice(xrange(2)) for _ in range(training_samples)]
clf = svm.SVC(kernel='linear', gamma=10, C=100)
clf.fit(X, y)
```

```python
prediction = clf.predict(features[:training_samples])
print sum(prediction == y)/(training_samples * 1.)
```
